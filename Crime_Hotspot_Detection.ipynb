```python
#Remove unwanted columns
df = df.drop(columns = ['Crm Cd 1','Crm Cd 2','Crm Cd 3','Crm Cd 4','Mocodes','Part 1-2','LAT','LON'])
```

```python
#Rename columns
renamed_colums = {'DR_NO':'ID','Date Rptd':'Reported_Date','DATE OCC':'Date_Occured','TIME OCC':'Time_Occured','AREA':'Area_Code','AREA NAME':'Area',
                  'Rpt Dist No':'District_No','Crm Cd':'Crime_Code','Crm Cd Desc':'Crime_Desc','Vict Age':'Victim_Age','Vict Sex':'Victim_Gender','Vict Descent':'Victim_Race',
                  'Premis Cd':'Crime_Loc_Code','Premis Desc':'Crime_Loc_Desc','Weapon Used Cd':'Weapon_Used_Code','Weapon Desc':'Weapon_Desc','LOCATION':'Location'}
df=df.rename(columns = renamed_colums)
```

```python
df.info()
```

```python
df['Area'].value_counts()
```

```python
#Replacing Code with Generic value
df['Victim_Gender'] = df['Victim_Gender'].replace(['X',np.nan, 'H', '-'],'Unknown')
df['Victim_Gender'] = df['Victim_Gender'].replace({'M':'Male', 'F':'Female'})
```

```python
#Handling NULL values
df['Victim_Race'].fillna('NA',inplace=True)
df['Crime_Loc_Desc'].fillna('NA',inplace=True)
df['Crime_Loc_Code'].fillna(0,inplace=True)
df['Weapon_Desc'].fillna('NA',inplace=True)
df['Weapon_Used_Code'].fillna(0,inplace=True)
df['Cross Street'].fillna('NA',inplace=True)
```

```python
df['Victim_Gender'].unique()
```

```python
df['Crime_Desc'].sample(10)
```

```python
#Remove invalid time occurances
df = df[~((df['Time_Occured'] > 24) & (df['Time_Occured'].astype(str).apply(len) == 2)) ]
```

```python
df['Date_Occured']=pd.to_datetime(df['Date_Occured'])
df['Reported_Date']=pd.to_datetime(df['Reported_Date'])
```

```python
df['Year_Occured'] = df['Date_Occured'].dt.year
df['Month_Occured'] = df['Date_Occured'].dt.month_name().str.slice(stop=3)
```

```python
df.head()
```

```python
df.Year_Occured.value_counts()
```

```python
#We have data only till Oct 2024 so we are excluding 2024 data
df = df[df['Year_Occured']!=2024]
```

#DATA VISUALISATION

**Distribution of Victims Age**

```python
#let's add a new variable named Vict_Age_Group and then build a histogram to see the distribution of victims' ages

age_ranges = [0, 17, 25, 34, 44, 54, 64, float('inf')]
age_labels = ['<18', '18-25', '26-34', '35-44', '45-54', '55-64', '65+']
df['Victim Age Group'] = pd.cut(df['Victim_Age'], bins=age_ranges, labels=age_labels)
```

```python
#as we can see from the histogram people of age 26-44 become victim more often then other

sns.histplot(data=df, x='Victim Age Group')
plt.show()
```

***This Histogram Plot showing the age of the victims. And also we can see that from the histogram people of age 26-44 become victim more often than others. ***

**Top 10 Areas where Most Crimes occur**

```python
plt.figure(figsize = (10,5))
sns.countplot(y='Area', data = df, order = df['Area'].value_counts().iloc[0:10].index)
plt.title('Top 10 Crime Areas')
plt.xlabel('Number of crimes')
plt.show()
```

**Crimes By Area**

**Below graph shows the Crimes happendend in the 21 Areas. Through this graph its shows the detailed about the crimes happend in the different placeses.**

```python
crimeByArea = df['Area'].value_counts().sort_index()
crimeCommonType = {}
for area in crimeByArea.keys():
    crimeArea = df[df['Area'] == area]['Crime_Desc'].value_counts()[:10]
    for crType in crimeArea.keys():
        if not crType in crimeCommonType:
            crimeCommonType[crType] = [area]
        else:
            crimeCommonType[crType].append(area)
    crimeArea = crimeArea.plot(kind = 'bar',title = "Crimes in " + area)
    plt.show()
```

**Crimes over the Year (2020 to 2024)**

```python
year_crimes_count = df['Year_Occured'].astype(str).value_counts().sort_index()
year_crimes_count
plt.figure(figsize=(12,6))
year_crimes_count.plot(kind = 'line')
plt.title('Occurrences of Crime From Year 2020-2024')
plt.ylabel('Number of crimes')
plt.xlabel('Year')
plt.show()
```

**This Graph shows that In Between 2020 to 2024 the crimes rate is increasing. From 2020 the graph is going upward and from 2022 its slowly going down. **

**Crimes Occurences By Month**

```python
month_crimes_count = df['Month_Occured'].astype(str).value_counts().sort_index()

months =  ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
month_crimes_count = month_crimes_count.reindex(months)
month_crimes_count
plt.figure(figsize=(10,4))
month_crimes_count.plot(kind = 'line')
plt.title('Occurrences of Crime by Month')
plt.ylabel('Number of crimes')
plt.xlabel('Month')
plt.show()
```

**This graph showing crimes occurences by month. And the graph shows the rating is like going upward and downward.But the end shows the graph going upward.**

**Top 10 Weapons used in Crime from 2020 to 2024**

```python
weapons_used= df[df['Weapon_Used_Code']>0]
weapons_used = weapons_used.groupby(["Weapon_Used_Code", "Weapon_Desc"]).size().reset_index(name="Number Of Crimes")
weapons_used = weapons_used.nlargest(10, "Number Of Crimes")
weapons_used.head()
```

**In the above table shows that the weapons and number of crimes. 159K People were killed by Strong Arm. And the 32k people were killed by Unknown Weapon etc.. and these are the detailde about the above table.**

```python
plt.figure(figsize=(13,7))
sns.barplot(data=weapons_used, x="Weapon_Used_Code", y="Number Of Crimes", palette="flare",order=weapons_used.sort_values(by="Number Of Crimes", ascending=False)["Weapon_Used_Code"])
plt.xlabel("Weapon Used Code",  fontsize=10)
plt.ylabel("Number Of Crimes",  fontsize=10)
plt.title("Top 10 Weapons Used In Crime", fontweight="bold", fontsize=12)
legend = [f"{weapon_code} - {weapon_desc}" for weapon_code, weapon_desc in zip (weapons_used["Weapon_Used_Code"], weapons_used["Weapon_Desc"])]
plt.legend(legend, title="Weapons Used Codes and Descriptions", bbox_to_anchor=(1, 1), loc="upper left")
plt.show()
```

**This graph shows the detailed about the table above. **

**Top 10 Crimes Reported in 2020 to 2024**

```python
crimes = df.groupby(["Crime_Code", "Crime_Desc"]).size().reset_index(name="Number Of Crimes")
crimes = crimes.nlargest(10, "Number Of Crimes")
crimes.head()
```

```python
plt.figure(figsize=(13,7))
sns.barplot(data=crimes, x="Crime_Code", y="Number Of Crimes", palette="crest",order=crimes.sort_values(by="Number Of Crimes", ascending=False)["Crime_Code"])
plt.xlabel("Crime Code",  fontsize=10)
plt.ylabel("Number Of Crimes",  fontsize=10)
plt.title("Top 10 Crimes Reported", fontweight="bold", fontsize=12)
legend = [f"{crime_code} - {crime_desc}" for crime_code, crime_desc in zip (crimes["Crime_Code"], crimes["Crime_Desc"])]
plt.legend(legend, title="Crime Codes and Descriptions", bbox_to_anchor=(1, 1), loc="upper left")
plt.show()
```

**The above graph shows the Crime codes and description.**

**Top 20 Premises where Crimes occured**

```python
top_20_premises = df.groupby(["Crime_Loc_Code", "Crime_Loc_Desc"]).size().reset_index(name="Number Of Crimes")
top_20_premises = top_20_premises.nlargest(20, "Number Of Crimes")
top_20_premises.head()
```

```python
plt.figure(figsize=(15,9))

# Bar plot creation
sns.barplot(data=top_20_premises, x="Crime_Loc_Code", y="Number Of Crimes", palette="rocket",
           order=top_20_premises.sort_values(by="Number Of Crimes", ascending=False)["Crime_Loc_Code"])
plt.xlabel("Crime_Loc_Code", fontweight="bold", fontsize=12)
plt.ylabel("Number Of Crimes", fontweight="bold", fontsize=12)
plt.title("Top 20 Premises Where Crimes Occured", fontweight="bold", fontsize=14)

# Legend creation and tweaking
legend = [f"{code} - {description}" for code, description in zip (top_20_premises["Crime_Loc_Code"], top_20_premises["Crime_Loc_Desc"])]
legend_obj = plt.legend(legend, title="Premise Codes And Descriptions", bbox_to_anchor=(1.05,1), loc="upper left")
title = legend_obj.get_title()
title.set_fontsize(12), title.set_fontweight("bold")

plt.show()
```

**This graphs shows the top 20 premises where crimes occured.**
